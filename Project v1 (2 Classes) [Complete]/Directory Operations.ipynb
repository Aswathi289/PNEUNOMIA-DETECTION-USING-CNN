# Folder Size

```python
def get_directory_size(directory):
    total_size = 0

    for dirpath, dirnames, filenames in os.walk(directory):
        for filename in filenames:
            file_path = os.path.join(dirpath, filename)
            total_size += os.path.getsize(file_path)

    size_in_bytes = total_size
    size_in_kilobytes = total_size / 1024
    size_in_megabytes = total_size / (1024 * 1024)
    size_in_gigabytes = total_size / (1024 * 1024 * 1024)

    return size_in_bytes, size_in_kilobytes, size_in_megabytes, size_in_gigabytes

directory_path = '/content/drive/MyDrive/S6/Mini Project/Project v2 (PNCT)/v2 Dataset 5.1.2 (replaced with COVID-19 Radiography Dataset)/val/TUBERCULOSIS'

size_bytes, size_kb, size_mb, size_gb = get_directory_size(directory_path)

print(f"Size: {size_bytes} bytes, {size_kb:.2f} KB, {size_mb:.2f} MB, {size_gb:.2f} GB")
```

#Directory Summation (images only)

```python
# To know the no. of images in each path of root folder

root_folder = '/content/drive/MyDrive/S6/Mini Project/Project v2 (PNCT)/v2 Dataset 5.1.2 (replaced with COVID-19 Radiography Dataset)/val'

for dirpath, dirnames, filenames in os.walk(root_folder):
    # Count the number of image files in each subfolder
    image_count = sum(1 for filename in filenames if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')))

    # Display the subfolder path and the corresponding image count
    print(f"Subfolder: {dirpath}")
    print(f"Number of Images: {image_count}")
    print()
```

# Copy

```python
# Specify the current location and the destination location for the copy
source = '/path/to/current/file.txt'
destination = '/path/to/new/location/file.txt'

# Copy the file to the new location
shutil.copy(source, destination)
```

```python
# Copy a directory and its contents recursively

src_dir = '/content/drive/MyDrive/S6/Mini Project/Project v2 (PNCT)/v2 Dataset 5.1.1 (replaced with COVID-19 Radiography Dataset)'
dest_dir = '/content/drive/MyDrive/S6/Mini Project/Project v2 (PNCT)/[copy] v2 Dataset 5.1.1 (replaced with COVID-19 Radiography Dataset)'

src_items = os.listdir(src_dir)
print("Number of elements in the source directory:", len(src_items))

shutil.copytree(src_dir, dest_dir)

dest_items = os.listdir(dest_dir)
print("Number of elements in the destination directory:", len(dest_items))
```
