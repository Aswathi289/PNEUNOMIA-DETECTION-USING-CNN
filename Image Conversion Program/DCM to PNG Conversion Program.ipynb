```python
# Whole Folder Conversion

# Locate the folder containing the DICOM files
dicom_folder = r'C:\Users\ACER\Desktop\Jupyter\Belarus Dataset Conversion\Belarus TB (extracted)\images'

# Create an output folder for the PNG images
output_folder = r'C:\Users\ACER\Desktop\Jupyter\Belarus Dataset Conversion\Belarus TB (converted png)'
os.makedirs(output_folder, exist_ok=True)

# Convert DICOM files to PNG
for filename in os.listdir(dicom_folder):
    if filename.endswith('.dcm'):
        # Read the DICOM file
        dcm_file = os.path.join(dicom_folder, filename)
        ds = pydicom.dcmread(dcm_file)

        # Extract the pixel array
        pixel_array = ds.pixel_array

        # Apply rescaling if necessary
        if 'Rescale Slope' in ds and 'Rescale Intercept' in ds:
            rescale_slope = ds.RescaleSlope
            rescale_intercept = ds.RescaleIntercept
            pixel_array = pixel_array * rescale_slope + rescale_intercept

        # Normalize pixel values to 8-bit range (0-255)
        pixel_array = np.interp(pixel_array, (pixel_array.min(), pixel_array.max()), (0, 255))

        # Convert to uint8 data type
        pixel_array = pixel_array.astype(np.uint8)

        # Create a PIL image from the pixel array
        image = Image.fromarray(pixel_array)

        # Save the image as PNG
        png_filename = os.path.splitext(filename)[0] + '.png'
        png_file = os.path.join(output_folder, png_filename)
        image.save(png_file)
```
